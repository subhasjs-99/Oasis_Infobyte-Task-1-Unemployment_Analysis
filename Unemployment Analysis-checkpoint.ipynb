```python
dataset = dataset.rename(columns={dataset.columns[0]:'State',dataset.columns[3]:'EUR',dataset.columns[4]:'EE', dataset.columns[5]:'ELPR',dataset.columns[2]:'FF',dataset.columns[1]:'date'})
dataset.head()
```

Renaming columns name for analyzing the data easily that will help us to using those feature efficiently. 
Convert "Region->State","Estimated Unemployment Rate (%)->EUR","Estimated Employed->EE","Frequency->FF",
        "Date->date","Estimated Labour Participation Rate (%)->ELPR".

```python
feature=['EUR','EE','ELPR']
dataset[feature].corr()
```

FIND OUT THE CO-RELATION BETWEEN 'EUR','EE' & 'ELPR'.
HERE WE UNDERSTAND NONE ARE DEPENDABLE TO EACH OTHER BEACUSE NONE OF CLOSE VALES WE ARE DETERMINE BETWEEN THESE FEATURES

```python
sn.heatmap(dataset[feature].corr(),annot=True)
plt.title("Heat map of Feature")
plt.show()
```

PLOT A HEATMAP TO GET A BETTER UNDERSTANDING OF THE CO-RELATION BETWEEN 'EUR','EE' & 'ELPR'.

```python
sn.pairplot(dataset[feature])
plt.show()
```

PLOT A PAIRPLOT TO GET A BETTER UNDERSTANDING THE RELATIONSHIP BETWEEN 'EUR','EE' & 'ELPR'.

```python
group_state=dataset.groupby(['State'])[['EUR','EE','ELPR']].mean().reset_index()
```

```python
group_state
```

UNDERSTAND EUR,EE & ELPR OF MEAN VALUE ACCORDING TO STATE

```python
group_area=dataset.groupby(['Area'])[['EUR','EE','ELPR']].mean().reset_index()
group_area
```

UNDERSTAND EUR,EE & ELPR OF MEAN VALUE ACCORDING TO AREA

```python
sn.boxplot(group_area['EUR'])
plt.title('Box plot of Estimated Unemployment Rate according to Area')
plt.show()
```

PLOTTING A BOXPLOT OF MEAN OF EUR ACCORDING TO AREA.
HERE WE DETERMINE MOST VALUE OF EUR < 12.5 & EUR > 11 AND THE MEDIAN VALUE BETWEEN 11.5-12.0

```python
sn.boxplot(group_area['EE'])
plt.title('Box plot of Estimated Employment according to Area')
plt.show()
```

PLOTTING A BOXPLOT OF MEAN OF EE ACCORDING TO AREA
PLOTTING A BOXPLOT OF MEAN OF EE ACCORDING TO AREA. HERE WE DETERMINE MOST VALUE OF EE < 0.9 & EE < 0.5 AND THE MEDIAN VALUE BETWEEN 0.7-0.8

```python
sn.boxplot(group_area['ELPR'])
plt.title('Box plot of Estimated Labour Employment Rate according to Area')
plt.show()
```

PLOTTING A BOXPLOT OF MEAN OF ELPR ACCORDING TO AREA
PLOTTING A BOXPLOT OF MEAN OF ELPR ACCORDING TO AREA. HERE WE DETERMINE MOST VALUE OF ELPR < 43.7 & EUR > 41.5 AND THE MEDIAN VALUE BETWEEN 43.0-42.5

```python
df1=dataset.groupby('State')[['EUR']].mean().reset_index()
sn.barplot(x='EUR',y='State',data=df1)
plt.title('Bar plot of Estimated Unemployment Rate according to State')
plt.show()
```

PLOTTING A BAR PLOT OF MEAN OF EUR ACCORING TO STATE
HERE WE DETERMINE UMEMPLOYMENT RATE IS MAXIMUM IN TRIPURA AND THEN HARYANA

```python
df2=dataset.groupby('State')['EE'].mean().reset_index()
sn.barplot(x='EE',y='State',data=df2)
plt.title('Bar plot of Estimated Employment according to State')
plt.show()
```

PLOTTING A BAR PLOT OF MEAN OF EE ACCORING TO STATE
HERE WE DETERMINE ESTIMATED EMPLOYMENT IS MAXIMUM IN UTTARPRADESH AND MAHARASTRA

```python
df3=dataset.groupby('State')['ELPR'].mean().reset_index()
sn.barplot(x='ELPR',y='State',data=df3)
plt.title('Bar plot of Estimated Labour Employment Rate according to State')
plt.show()
```

PLOTTING A BAR PLOT OF MEAN OF ELPR ACCORING TO STATE
HERE WE DETERMINE ESTIMATED LABOUR EMPLOYMENT RATE IS MAXIMUM IN TRIPURA AND THEN MEGHALAYA

```python
df4=dataset.groupby('Area')['EUR'].mean().reset_index()
sn.barplot(x='Area',y='EUR',data=df4)
plt.title('Bar plot of Estimated Unemployment Rate according to Area')
plt.show()
```

PLOTTING A BARPLOT OF EUR OF MEAN ACCORDING TO AREA
HERE WE UNDERSTAND THAT ESTIMATED UNEMPLOYMENT RATE IS MORE IN URBAN AREAS

```python
df5=dataset.groupby('Area')['EE'].mean().reset_index()
sn.barplot(x='Area',y='EE',data=df5)
plt.title('Bar plot of Estimated Employment according to Area')
plt.show()
```

PLOTTING A BARPLOT OF EE OF MEAN ACCORDING TO AREA
HERE WE UNDERSTAND THAT ESTIMATED EMPLOYMENT IS MORE IN RURAL AREAS

```python
df6=dataset.groupby('Area')['ELPR'].mean().reset_index()
sn.barplot(x='Area',y='ELPR',data=df6)
plt.title('Bar plot of Estimated Labour Employment Rate according to Area')
plt.show()
```

PLOTTING A BARPLOT OF ELPR OF MEAN ACCORDING TO AREA
HERE WE UNDERSTAND THAT ESTIMATED LABOUR EMPLOYMENT RATE IS MORE IN RURAL AREAS
